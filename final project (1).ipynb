# Handling outliers

```python
adult.boxplot(figsize=(10,5))
plt.show()
```

```python
numeric_columns = ['age','fnlwgt', 'education_level', 'capital_gain', 'capital_loss', 'hours_per_week']

for col in numeric_columns:
    
    Q1 = adult[col].quantile(0.25)
    Q3 = adult[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    adult[col] = adult[col].clip(lower_bound, upper_bound)
```
